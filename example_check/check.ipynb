```python
import cv2
import os

img_paths = [
    './00000-infrared.png',  # 红外图
    './00000-visible.png'  # 可见光图
]

for path in img_paths:
    img = cv2.imread(path, cv2.IMREAD_UNCHANGED)

    if img is None:
        print(f'{path} 读取失败')
        continue

    if img.ndim == 2:
        channels = 1
    else:
        channels = img.shape[2]

    print(f'{os.path.basename(path)}: shape={img.shape}, channels={channels}')
```

```python
from PIL import Image

img_paths = [
    './00000-infrared.png',  # 红外图
    './00000-visible.png'  # 可见光图
]

for path in img_paths:
    img = Image.open(path)
    print(f'{path}: mode={img.mode}, channels={len(img.getbands())}')
```

```python
import cv2
import numpy as np

img = cv2.imread('./00000-infrared.png', cv2.IMREAD_UNCHANGED)

b, g, r = cv2.split(img)

print(
    np.all(b == g),
    np.all(g == r)
)
```
